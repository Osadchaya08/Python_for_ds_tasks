__Формулювання бізнес-питань__

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

__1. Які продукти є найбільш популярними серед покупців (визначаємо популярність за категоріями товарів):__

```python
# Підрахунок кількості продажів за категоріями
sales_by_category = df['Category'].value_counts().reset_index()
sales_by_category.columns = ['Category', 'Total Sales']

# Сортування за кількістю продажів у спадаючому порядку
sales_by_category = sales_by_category.sort_values(by='Total Sales', ascending=False)

# Виведення результатів
print("Найбільш популярні категорії товарів:")
print(sales_by_category)
```

Висновок: Найбільш популярна категорія товарів "Bakery & Breakfast", загальний обсяг якої станочить 6770 продажів.

__2. Яка категорія товарів приносить найбільший дохід?__

```python
# Додаткові обчислення: доход для кожного товару (ціна * кількість)
df['Total Revenue'] = df['Price'] * df['QTY']

# Групування за категорією товарів і підсумовування доходів
category_revenue = df.groupby('Category')['Total Revenue'].sum().reset_index()

# Сортування за загальним доходом у спадаючому порядку
category_revenue = category_revenue.sort_values(by='Total Revenue', ascending=False)

# Виведення результатів
print("Категорії товарів за найбільшим доходом:")
print(category_revenue)
```

Висновок: Категорія, яка найбільше приносить прибутку "Beauty", загальна виручка становить 81612.17. 

```python
plt.figure(figsize=(10, 6))
plt.bar(category_revenue['Category'], category_revenue['Total Revenue'], color='skyblue')
plt.xlabel('Категорія товарів')
plt.ylabel('Загальний дохід')
plt.title('Загальний дохід по категоріях товарів')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show();
```

__3. Які товарні категорії є найбільш стабільними у продажах?__

```python
# Групування за товарною категорією і підрахунок середньомісячних продажів
monthly_sales = df.groupby('Category')['QTY'].mean()

# Розрахунок коефіцієнта варіації
std_dev = df.groupby('Category')['QTY'].std()
coeff_variation = std_dev / monthly_sales * 100

# Округлення до двох знаків після коми
coeff_variation_rounded = coeff_variation.round(2)

# Відсортувати за коефіцієнтом варіації від найменшого до найбільшого
coeff_variation_sorted = coeff_variation_rounded.sort_values(ascending=True)

print("Найбільш стабільні товарні категорії за коефіцієнтом варіації:")
print(coeff_variation_sorted)
```

__4.Які регіони мають найвищий і найнижчий рівень продажів?__

```python
# Групування за регіоном і розрахунок сумарних продажів
sales_by_region = df.groupby('City')['Amount'].sum()

# Знайдення регіону з найвищим рівнем продажів
max_sales_region = sales_by_region.idxmax()
max_sales_amount = sales_by_region[max_sales_region]

# Знайдення регіону з найнижчим рівнем продажів
min_sales_region = sales_by_region.idxmin()
min_sales_amount = sales_by_region[min_sales_region]

# Вивід результатів
print(f"Місто з найвищим рівнем продажів: {max_sales_region}, сума продажів: {max_sales_amount}")
print(f"Місто з найнижчим рівнем продажів: {min_sales_region}, сума продажів: {min_sales_amount}")
```

__5. Які продукти мають найбільшу сезонність в продажах?__

```python
# Групування за продуктом і розрахунок середньомісячних продажів
monthly_sales = df.groupby('Category')['QTY'].mean()

# Розрахунок коефіцієнта варіації
std_dev = df.groupby('Category')['QTY'].std()
coeff_variation = std_dev / monthly_sales * 100

# Відсортувати за коефіцієнтом варіації (сезонністю) з округленням до двох знаків
coeff_variation_sorted = coeff_variation.sort_values(ascending=False).round(2).head(5)  # Вибрати топ-5 продуктів

# Побудова графіків за допомогою plt.subplot
plt.figure(figsize=(12, 8))

for i, (product, variation) in enumerate(coeff_variation_sorted.items(), start=1):
    plt.subplot(2, 3, i)
    df[df['Category'] == product].groupby('Month')['QTY'].sum().plot(kind='line', marker='o', color='blue')
    plt.title(f'Продукт: {product}')
    plt.xlabel('Місяць')
    plt.ylabel('Сума продажів')
    plt.grid(True)

plt.tight_layout()
plt.show()

# Вивід найбільш сезонних продуктів
print("Продукти з найбільшою сезонністю в продажах:")
print(coeff_variation_sorted)
```

__6. Які товарні позиції найбільше впливають на загальні прибутки компанії?__

```python
# Групування за товарною позицією і розрахунок суми прибутків за кожну позицію
total_profit = df.groupby('Items')['Profit Margin'].sum()

# Відсортувати за сумою прибутків у спадному порядку
total_profit_sorted = total_profit.sort_values(ascending=False).head(10)  # Вибрати топ-10 товарних позицій

# Вивід результатів
print("Товарні позиції, які найбільше впливають на загальні прибутки компанії:")
print(total_profit_sorted)
```

Висновок: саме позиція "Ashai Lager Beer" найбільше приносить прибуток компанії.

__7.Чи є різниця у середньому чеку між чоловіками та жінками?__

```python
# Групування за статтю і обчислення середнього чеку
avg_check = df.groupby('Gender')['Amount'].mean()

# Побудова кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(avg_check, labels=avg_check.index, autopct='%1.1f%%', startangle=140)
plt.title('Середній чек за статтю')
plt.legend(title='Стать')
plt.show();
```

Висновок:цей підхід дозволяє візуально порівняти середній чек між різними групами (чоловіки та жінки) за допомогою кругової діаграми з автоматичним виведенням відсотків.В жінок середній чек більший, але не суттєво.
